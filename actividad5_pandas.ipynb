```python
# ============================================================
# Configuración para mostrar TODO el output en Jupyter Notebook
# ============================================================
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# ============================================================
# Dataset base
# ============================================================
datos = {
    'Nombre':  ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'Edad':    [25, 30, 22, None, 28],
    'Ciudad':  ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'Ingreso': [3000, 4500, 2800, 5000, None]
}

df = pd.DataFrame(datos)
df.to_csv('actividad_semana5.csv', index=False)
print("✅ Dataset base creado y guardado como 'actividad_semana5.csv'\n")

# ============================================================
# EJERCICIO 1: Series — crear y operar
# ============================================================
print("============================")
print("EJERCICIO 1: Series — crear y operar")
print("============================\n")

serie_lista = pd.Series([10, 20, 30, 40, 50])
print("📘 Series desde lista:")
print(serie_lista, "\n")

serie_dicc = pd.Series({'a': 100, 'b': 200, 'c': 300})
print("📗 Series desde diccionario:")
print(serie_dicc, "\n")

print("➡️ Valor en índice 2:", serie_lista[2], "\n")

serie_lista[2] = 35
print("📙 Serie modificada:")
print(serie_lista, "\n")

print("📒 Serie multiplicada por 2:")
print(serie_lista * 2, "\n")

# ============================================================
# EJERCICIO 2: DataFrame — crear y explorar
# ============================================================
print("============================")
print("EJERCICIO 2: DataFrame — crear y explorar")
print("============================\n")

data = {
    'Producto': ['Pan', 'Leche', 'Queso'],
    'Precio': [1500, 3500, 8000],
    'Cantidad': [10, 20, 5]
}
df2 = pd.DataFrame(data)
print("📊 DataFrame creado:")
print(df2, "\n")

print("📋 Tipos de datos:")
print(df2.dtypes, "\n")

# ============================================================
# EJERCICIO 3: Inspeccionar y explorar
# ============================================================
print("============================")
print("EJERCICIO 3: Inspeccionar y explorar")
print("============================\n")

print("🔍 Primeras filas:")
print(df.head(3), "\n")

print("🔎 Últimas filas:")
print(df.tail(2), "\n")

print("ℹ️ Información general:")
df.info()
print("\n📏 Dimensiones (filas, columnas):", df.shape, "\n")

print("📈 Resumen estadístico:")
print(df.describe(), "\n")

# ============================================================
# EJERCICIO 4: Acceder a columnas y filas
# ============================================================
print("============================")
print("EJERCICIO 4: Acceder a columnas y filas")
print("============================\n")

print("📌 Columna 'Nombre':")
print(df['Nombre'], "\n")

print("📌 Fila con índice 1 usando loc:")
print(df.loc[1], "\n")

print("📌 Fila con índice 1 usando iloc:")
print(df.iloc[1], "\n")

print("📌 Valor específico (fila 1, columna Edad):", df.loc[1, 'Edad'], "\n")

# ============================================================
# EJERCICIO 5: Operaciones básicas
# ============================================================
print("============================")
print("EJERCICIO 5: Operaciones básicas")
print("============================\n")

df['Edad'] = df['Edad'].fillna(0) + 5
df['Ingreso_anual'] = df['Ingreso'].fillna(0) * 12
print("🧮 DataFrame con edad incrementada y columna 'Ingreso_anual':")
print(df, "\n")

# ============================================================
# EJERCICIO 6: Filtrar datos
# ============================================================
print("============================")
print("EJERCICIO 6: Filtrar datos")
print("============================\n")

print("🔹 Filas con Edad > 30:")
print(df[df['Edad'] > 30], "\n")

print("🔸 Filas donde Ciudad es 'Madrid' o 'Lima':")
print(df[df['Ciudad'].isin(['Madrid', 'Lima'])], "\n")

# ============================================================
# EJERCICIO 7: Manejo de valores faltantes
# ============================================================
print("============================")
print("EJERCICIO 7: Manejo de valores faltantes")
print("============================\n")

print("🔍 Detección de valores nulos:")
print(df.isna(), "\n")

print("📊 Conteo de valores faltantes por columna:")
print(df.isna().sum(), "\n")

df['Edad'] = df['Edad'].fillna(0)
df['Ciudad'] = df['Ciudad'].fillna('Desconocido')
df['Ingreso'] = df['Ingreso'].fillna(df['Ingreso'].median())
print("✅ DataFrame con valores faltantes rellenados:")
print(df, "\n")

# ============================================================
# EJERCICIO 8: Leer y guardar datos
# ============================================================
print("============================")
print("EJERCICIO 8: Leer y guardar datos")
print("============================\n")

df_leido = pd.read_csv('actividad_semana5.csv')
print("📂 CSV leído:")
print(df_leido, "\n")

df_leido[['Nombre', 'Edad', 'Ciudad']].to_csv('actividad5_seleccion.csv', index=False)
print("✅ Archivo 'actividad5_seleccion.csv' guardado con columnas seleccionadas\n")

# ============================================================
# EJERCICIO 9: Ejercicio integrado
# ============================================================
print("============================")
print("EJERCICIO 9: Ejercicio integrado")
print("============================\n")

df_integrado = df.copy()
df_integrado['Edad'] += 5
df_integrado['Ingreso_anual'] = df_integrado['Ingreso'].fillna(0) * 12

personas_filtradas = df_integrado[df_integrado['Ingreso_anual'] > 36000].sort_values(by='Ingreso_anual', ascending=False)
personas_filtradas.to_csv('personas_filtradas.csv', index=False)
print("✅ Archivo 'personas_filtradas.csv' creado con datos filtrados y ordenados\n")

print("📈 Resultado final:")
print(personas_filtradas)
```
